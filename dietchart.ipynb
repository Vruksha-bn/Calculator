```python
import pandas as pd

# Load the dataset (replace 'actual_dataset_file.csv' with the correct file name)
df = pd.read_csv('bmi.csv')

# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Check the column names in the dataset
print(df.columns)
```

```python
# Replace 'BMI' with the actual column name from the dataset
df['Bmi'].fillna(df['Bmi'].mean(), inplace=True)
```

```python
# Remove any leading or trailing spaces from the column names
df.columns = df.columns.str.strip()

# Now use the correct 'BMI' column name
df['Bmi'].fillna(df['Bmi'].mean(), inplace=True)
```

```python
# Check the column names first
print(df.columns)

# Assuming the actual column name is 'bmi' (lowercase), use this:
df['Bmi'].fillna(df['Bmi'].mean(), inplace=True)
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Height'].fillna(df['Height'].median(), inplace=True)

# Check if missing values are handled
print(df.isnull().sum())
```

```python
# Replace inplace=True with assignment
df['Bmi'] = df['Bmi'].fillna(df['Bmi'].mean())
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Height'] = df['Height'].fillna(df['Height'].median())
```

```python
# Use .loc for assigning values
df.loc[:, 'Bmi'] = df['Bmi'].fillna(df['Bmi'].mean())
df.loc[:, 'Age'] = df['Age'].fillna(df['Age'].median())
df.loc[:, 'Height'] = df['Height'].fillna(df['Height'].median())
```

```python
# Option 1: Avoid inplace=True by direct assignment
df['Bmi'] = df['Bmi'].fillna(df['Bmi'].mean())
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Height'] = df['Height'].fillna(df['Height'].median())

# Or Option 2: Use .loc for assignment
df.loc[:, 'Bmi'] = df['Bmi'].fillna(df['Bmi'].mean())
df.loc[:, 'Age'] = df['Age'].fillna(df['Age'].median())
df.loc[:, 'Height'] = df['Height'].fillna(df['Height'].median())
```

```python
# Drop duplicate rows
df_cleaned = df.drop_duplicates()

# Check if duplicates were removed
print(f"Number of duplicates: {df.duplicated().sum()}")
```

```python
# Convert Age to integer if needed
df['Age'] = df['Age'].astype(int)

# Convert Height, Weight, and BMI to float if they are not
df['Height'] = df['Height'].astype(float)
df['Weight'] = df['Weight'].astype(float)
df['Bmi'] = df['Bmi'].astype(float)

# Check the data types to ensure they're correct
print(df.dtypes)
```

```python
# Define thresholds for outliers based on domain knowledge (example values for illustration)
df_cleaned = df[(df['Height'] > 50) & (df['Height'] < 250)]  # Height in cm
df_cleaned = df_cleaned[(df['Weight'] > 20) & (df['Weight'] < 300)]  # Weight in kg
df_cleaned = df_cleaned[(df['Bmi'] > 10) & (df['Bmi'] < 60)]  # Reasonable BMI range

# Check after removing outliers
print(df_cleaned.describe())
```

```python
# Recalculate BMI using height in meters
df['Height_m'] = df['Height'] / 100  # Convert height from cm to meters
df['Bmi_calculated'] = df['Weight'] / (df['Height_m'] ** 2)

# Compare calculated BMI with the existing BMI
print(df[['Bmi', 'Bmi_calculated']].head())
```

```python
# Save the cleaned dataset to a CSV file
df_cleaned.to_csv('cleaned_bmi_dataset.csv', index=False)
```
